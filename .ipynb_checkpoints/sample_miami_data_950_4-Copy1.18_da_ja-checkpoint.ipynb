```python
# Let's import our dependencies. 
import pandas as pd
import requests
import json
import numpy as np

import matplotlib.pyplot as plt

# Import the API key
from config_br import geoapify_key
```

```python
print(geoapify_key)
```

```python
sample_miami_df = pd.read_csv('sample_miami_df.csv')
sample_miami_df.head()
```

```python
school_lat_list = []
school_long_list =[]
school_count =[]
school_zip_list =[]

for index,row in sample_miami_df.iterrows():
    categories = "education.school"
    radius = 1000
    filters = f'circle:{row["LONGITUDE"]},{row["LATITUDE"]},{radius}'
    bias = f'proximity:{row["LONGITUDE"]},{row["LATITUDE"]}'
    limit = 100

    # set up a parameters dictionary
    params = {
        "categories":categories,
        "bias":bias,
        "apiKey":geoapify_key,    
        "limit":limit,
        "filter":filters,
        "bias":bias,
    }
    base_url = "https://api.geoapify.com/v2/places"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)   
    # print the response url, avoid doing for public github repos in order to avoid exposing key
    print(response.url)   
    school_lat_list.append(row["LATITUDE"])
    school_long_list.append(row["LONGITUDE"])
    places_data = response.json()
    school_count.append(len(places_data["features"]))
    try:
        school_zip_list.append(places_data["features"][0]["properties"]["postcode"])
    except:
        print("No schools in this area")
        school_zip_list.append(0)
```

```python
# Put schools data into a dataframe. 

schools_count_df = pd.DataFrame( {
    "LATITUDE": school_lat_list,
    "LONGITUDE": school_long_list,
    "school count" : school_count,
    })

schools_count_df
```

```python
# Now we're looking for the number of hospitals within 5,000m of each house. 

hosp_lat_list = []
hosp_long_list =[]
hosp_count =[]
hosp_zip_list =[]

for index,row in sample_miami_df.iterrows():
    categories = "healthcare.hospital"
    radius = 5000
    filters = f'circle:{row["LONGITUDE"]},{row["LATITUDE"]},{radius}'
    bias = f'proximity:{row["LONGITUDE"]},{row["LATITUDE"]}'
    limit = 100

    # set up a parameters dictionary
    params = {
        "categories":categories,
        "bias":bias,
        "apiKey":geoapify_key,    
        "limit":limit,
        "filter":filters,
        "bias":bias,
    }
    base_url = "https://api.geoapify.com/v2/places"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)   
    # print the response url, avoid doing for public github repos in order to avoid exposing key
    print(response.url)   
    hosp_lat_list.append(row["LATITUDE"])
    hosp_long_list.append(row["LONGITUDE"])
    places_data = response.json()
    hosp_count.append(len(places_data["features"]))
    try:
        hosp_zip_list.append(places_data["features"][0]["properties"]["postcode"])
    except:
        print("No hospitals in this area")
        hosp_zip_list.append(0)
```

```python
# Put hospitals data into a dataframe. 

hosp_count_df = pd.DataFrame( {
    "LATITUDE": hosp_lat_list,
    "LONGITUDE": hosp_long_list,
    "hosp count" : hosp_count,
    })

hosp_count_df
```

```python
# Now we can do our for loop to get all the restaurants within 1,000m of each house. 

rest_lat_list = []
rest_long_list =[]
rest_count =[]
rest_zip_list =[]

for index,row in sample_miami_df.iterrows():
    categories = "catering.restaurant"
    radius = 1000
    filters = f'circle:{row["LONGITUDE"]},{row["LATITUDE"]},{radius}'
    bias = f'proximity:{row["LONGITUDE"]},{row["LATITUDE"]}'
    limit = 100

    # set up a parameters dictionary
    params = {
        "categories":categories,
        "bias":bias,
        "apiKey":geoapify_key,    
        "limit":limit,
        "filter":filters,
        "bias":bias,
    }
    base_url = "https://api.geoapify.com/v2/places"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)   
    # print the response url, avoid doing for public github repos in order to avoid exposing key
    print(response.url)   
    rest_lat_list.append(row["LATITUDE"])
    rest_long_list.append(row["LONGITUDE"])
    places_data = response.json()
    rest_count.append(len(places_data["features"]))
    try:
        rest_zip_list.append(places_data["features"][0]["properties"]["postcode"])
    except:
        print("No restaurants in this area")
        rest_zip_list.append(0)
```

```python
# Put restaurants data into a dataframe. 

rest_count_df = pd.DataFrame( {
    "LATITUDE": rest_lat_list,
    "LONGITUDE": rest_long_list,
    "rest count" : rest_count,
    })

rest_count_df
```
