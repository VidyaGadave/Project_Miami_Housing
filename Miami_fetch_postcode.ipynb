```python
# Let's import our dependencies. 
import pandas as pd
import requests
import json
import numpy as np

# Import the API key
from config import geoapify_key  
```

```python
# Read our CSV file into pandas dataframe. 

miami_df = pd.read_csv('sample_miami_df.csv')
miami_df.head()
```

```python
house_zip_list =[]

for index,row in miami_df.iterrows():
    params = {
        "lat":row["LATITUDE"],
        "lon":row["LONGITUDE"],
        "apiKey":geoapify_key    
        
    }
    
    base_url = "https://api.geoapify.com/v2/place-details?PARAMS"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)   
    places_data = response.json()
    # print the response url, avoid doing for public github repos in order to avoid exposing key
    print(response.url)   
    try:
         house_zip_list.append(places_data["features"][0]["properties"]["postcode"])
    except:
         print("No address found")
         house_zip_list.append(0)
```

```python
post_code = pd.Series(house_zip_list)
miami_df['post_code'] = post_code
miami_df.head(5)

miami_df.to_csv("miami_data_enriched.csv",index=False)
```
